```python
import pandas as pd
import requests
```

```python

# Étape 1: Charger les données de test
df_test = pd.read_csv('reconstituted_test_sampled.csv')
```

```python
# df_test.head()
```

```python
# Étape 2: Exclure la première colonne (l'ID)
df_sample = df_test.iloc[:10, 1:]  # Prendre les 10 premières lignes, en excluant la première colonne
```

```python
# df_sample
```

```python
data = df_sample.to_dict(orient='records')
# data
```

```python

# Étape 3: Envoyer une requête POST à l'API
url = "http://127.0.0.1:8000/predict"
# url = "https://scorecredit-93521a3704b4.herokuapp.com/predict"
response = requests.post(url, json={"data": data})
```

```python

# Étape 4: Vérifier et afficher la réponse
if response.status_code == 200:
    print("Réponse de l'API:", response.json())
else:
    print(f"Erreur {response.status_code}: {response.text}")
```
